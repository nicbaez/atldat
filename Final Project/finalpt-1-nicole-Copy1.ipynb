```python
import pandas as pd
import numpy as np 
import statsmodels.formula.api as sm
from sklearn.linear_model import LinearRegression 
import scipy, scipy.stats
import datetime

def read_csv_file(path):
   df = pd.read_csv(path) 
   return(df)

def get_date_from_string(date_obj): 
    return datetime.datetime.strptime(date_obj,"%Y-%m-%d").date()

path = "./autopython.csv"
```

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt

# this allows plots to appear directly in the notebook
%matplotlib inline
```

```python
table = read_csv_file(path)
```

```python
table["Date"] = table.Date.apply(get_date_from_string)
```
